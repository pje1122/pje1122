```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import urllib.request
from sklearn.cluster import KMeans
```

```python
#링크 받고 사진 리턴함
def url_to_image(url):
  resp = urllib.request.urlopen(url)
  img = np.asarray(bytearray(resp.read()), dtype='uint8')
  img = cv2.imdecode(img, cv2.IMREAD_COLOR)
  return img
```

```python
#사진에서 RGB 추출
def image_to_data(img):
  img = url_to_image(url)
  cv2_imshow(img) #사진 보여주기
  #print(img.shape) #(가로픽셀수, 세로픽셀수, 채널수)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #bgr을 rgb로
  img= img.reshape((img.shape[0] * img.shape[1], 3)) # height, width 통합  #직렬화 해준셈.
  #print(img.shape) #(직렬화한 픽셀수 , 채널수)
  return img
```

```python
def data_to_center(img):
  k = 1 # 블록화 비슷한 거인듯 (한 사진을 더 블록화 하진 않으려고 값 1로 두겠음)
  clt = KMeans(n_clusters = k)
  clt.fit(img)
  for center in clt.cluster_centers_:
      print(center)
  #출력된 '이미지의 평균 RGB값'을 아래 사이트에서 찾아보면 얼추 그런 톤임 확인 가능
  #https://www.rapidtables.org/ko/web/color/RGB_Color.html
  return center
```

```python
#메인코드
for screenshots_url in screenshots_urls:
    print(screenshots_url)
    url = screenshots_url
    img = url_to_image(url)
    img = image_to_data(img)
    center = data_to_center(img)
```
